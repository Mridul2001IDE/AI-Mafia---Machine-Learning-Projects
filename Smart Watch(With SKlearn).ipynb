```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df_x=pd.read_csv('./Downloads/Linear_X_Train.csv')
df_y=pd.read_csv('./Downloads/Linear_Y_Train.csv')
x=df_x.values.reshape((-1,1))
y=df_y.values.reshape((-1,1))
X=x
Y=(y-y.mean())/y.std()
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(X,Y)
```

```python
test=pd.read_csv('./Downloads/Linear_X_Test.csv')
testx=test.values.reshape((-1,1))
```

```python
output=model.predict(testx)
```

```python
print(model.intercept_)
```

```python
print(model.coef_)
```

```python
model.score(X,Y)
```

```python
plt.scatter(X,Y,label="Training Data")
plt.plot(testx,output,color="red",label="Predicted Data")
plt.legend()
plt.show()
```
