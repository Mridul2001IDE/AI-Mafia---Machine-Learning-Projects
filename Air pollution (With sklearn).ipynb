```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('./Downloads/Train.csv')
d=df.values
f1=d[:,0].reshape((-1,1))
f2=d[:,1].reshape((-1,1))
f3=d[:,2].reshape((-1,1))
f4=d[:,3].reshape((-1,1))
f5=d[:,4].reshape((-1,1))
y=d[:,5].reshape((-1,1))
Y=(y-y.mean())/y.std()
```

```python
plt.scatter(f1,Y,label='f1')
plt.scatter(f2,Y,label='f2')
plt.scatter(f3,Y,label='f3')
plt.scatter(f4,Y,label='f4')
plt.scatter(f5,Y,label='f5')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model2=LinearRegression()
model3=LinearRegression()
model4=LinearRegression()
model5=LinearRegression()
model1.fit(f1,Y)
model2.fit(f2,Y)
model3.fit(f3,Y)
model4.fit(f4,Y)
model5.fit(f5,Y)
```

```python
test=pd.read_csv('./Downloads/Test.csv')
t=test.values
tf1=t[:,0].reshape((-1,1))
tf2=t[:,1].reshape((-1,1))
tf3=t[:,2].reshape((-1,1))
tf4=t[:,3].reshape((-1,1))
tf5=t[:,4].reshape((-1,1))
```

```python
output1=model1.predict(tf1)
output2=model2.predict(tf2)
output3=model3.predict(tf3)
output4=model4.predict(tf4)
output5=model5.predict(tf5)
```

```python
print("For feature1 {} is intercept and {} is coefficient and {} is score".format(model1.intercept_,model1.coef_,model1.score(f1,Y)))
print("For feature2 {} is intercept and {} is coefficient and {} is score".format(model2.intercept_,model2.coef_,model2.score(f2,Y)))
print("For feature3 {} is intercept and {} is coefficient and {} is score".format(model3.intercept_,model3.coef_,model3.score(f3,Y)))
print("For feature4 {} is intercept and {} is coefficient and {} is score".format(model4.intercept_,model4.coef_,model4.score(f4,Y)))
print("For feature5 {} is intercept and {} is coefficient and {} is score".format(model5.intercept_,model5.coef_,model5.score(f5,Y)))
```
